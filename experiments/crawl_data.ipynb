```python
from os import listdir
from os.path import isfile, join
onlyfiles = ["./data/" + f for f in listdir("./data/") if isfile(join("./data/", f))]
if (len(onlyfiles) > 0):
    concatted_csv = pd.read_csv(onlyfiles[0])
    for i in range(1, len(onlyfiles)):
        concatted_csv = pd.concat([concatted_csv, pd.read_csv(onlyfiles[i])])
print(len(concatted_csv))
concatted_csv = concatted_csv.drop_duplicates(subset="CELEX Number")
print(len(concatted_csv))
concatted_csv.to_csv("./concatenated-all.csv")
```
